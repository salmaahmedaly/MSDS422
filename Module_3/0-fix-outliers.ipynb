```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../../data_sets/SA_clean_HMEQ_Loss.csv')
```

```python
df.head()
```

```python
def cap_vars (df, var):
    upper_cap = df[var].mean() + (3*df[var].std())
    df[f'O_{var}'] = np.where(df[var]> upper_cap, upper_cap, df[var])
    df[f'O_Flag_{var}'] = np.where(df[var]> upper_cap, 1, 0) 
    df= df.drop(columns= var)
    return df
```

```python
fix_cols = [i for i in df.columns if i.startswith('IMP') and i != 'IMP_TARGET_LOSS_AMT' ]
```

```python
for var in fix_cols:
    df = cap_vars (df= df, var= var)
```

```python
df.to_csv('../../data_sets/SA_clean_O_fixed_HMEQ_Loss.csv')
```
